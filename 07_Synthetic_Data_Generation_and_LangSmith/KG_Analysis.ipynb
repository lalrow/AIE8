```python
import json
from pathlib import Path

# Path to your KG JSON file (adjust if needed)
kg_path = Path("usecase_data_kg.json")

with open(kg_path, "r") as f:
    kg = json.load(f)

nodes = kg.get("nodes", [])
print(f"✅ Loaded {len(nodes)} nodes from {kg_path}")
```

```python
for i, node in enumerate(nodes[:3]):
    props = node["properties"]
    meta = props.get("document_metadata", {})
    print(f"\n--- Node {i+1} ---")
    print("Page:", meta.get("page"))
    print("Title:", meta.get("title"))
    print("Chunk length:", len(props.get("page_content", "")))
    print("Preview:", props.get("page_content", "")[:120], "...")
```

```python
lengths = [len(n["properties"]["page_content"]) for n in nodes]

print("\n📊 Chunk Stats")
print("Average length:", sum(lengths) / len(lengths) if lengths else 0)
print("Max length:", max(lengths) if lengths else 0)
print("Min length:", min(lengths) if lengths else 0)
```

```python
empty_nodes = [n for n in nodes if not n["properties"]["page_content"].strip()]
print(f"Empty nodes: {len(empty_nodes)}")
```

```python
import pandas as pd

rows = []
for n in nodes:
    props = n["properties"]
    meta = props.get("document_metadata", {})
    rows.append({
        "page": meta.get("page"),
        "title": meta.get("title"),
        "chunk_length": len(props.get("page_content", "")),
        "preview": props.get("page_content", "")[:80]
    })

df = pd.DataFrame(rows)
df.to_csv("KG_Summary.csv", index=False)
print("📂 Saved summary to KG_Summary.csv")
```
